### Section 4 — Appointment Analysis

**Objective**

Analyze appointment outcomes, scheduling trends, cancellations, and no-show rates to assess operational efficiency.

#### Business Question 15
**Manager's Question**

How many appointments are Completed, Cancelled, Scheduled, and No-show?

```python

query = """
SELECT status , count(*)
FROM appointments
GROUP BY status;
"""
pd.read_sql(query, conn) 
```

**Observation** : 
The appointment distribution is relatively balanced across all four statuses. However, the number of **No-show appointments (52)** is slightly higher than the number of **Completed appointments (46)**. This may indicate a potential patient attendance issue, suggesting the hospital could benefit from reminder calls, SMS notifications, or appointment confirmation systems to reduce no-shows.

#### Business Question 16
**Manager's Question**

How many appointments were scheduled each day?

```python

query = """
SELECT appointment_date,count (*)
FROM appointments
GROUP BY appointment_date
ORDER BY appointment_date;
"""
pd.read_sql(query, conn) 
```

**Observation** : 
The query calculates the total number of appointments scheduled for each day.

**Business Insight** : 
Daily appointment trends help hospital administrators identify busy periods and allocate resources more efficiently.

#### Business Question 17
**Manager's Question**

What percentage of appointments are No-show?

```python
query = """
SELECT 
sum(CASE WHEN status='No-show' THEN 1 ELSE 0 END) * 100.0 / count(*) AS no_show_percentage
FROM appointments;
"""
pd.read_sql(query, conn)
```

**Observation** : 
The query calculates the percentage of appointments where patients did not attend.

**Business Insight** : 
A high no-show rate can negatively affect operational efficiency and may indicate the need for appointment reminder systems.

#### Business Question 18
**Manager's Question**

No-show % by doctor

```python
query = """
SELECT doctors.doctor_id , doctors.first_name,doctors.last_name,
sum(CASE WHEN appointments.status='No-show' THEN 1 ELSE 0 END) * 100.0 / count(*) AS no_show_percentage
FROM doctors
LEFT JOIN appointments
ON doctors.doctor_id = appointments.doctor_id
GROUP BY doctors.doctor_id , doctors.first_name,doctors.last_name
ORDER BY no_show_percentage DESC;
"""
pd.read_sql(query, conn)
```

**Observation** : 
The query calculates the percentage of appointments marked as No-show for each doctor by comparing missed appointments against the total appointments handled by that doctor.

**Business Insight** : 
Comparing no-show rates across doctors helps identify scheduling patterns and patient attendance behavior. Doctors with consistently higher no-show percentages may benefit from improved appointment reminder systems, scheduling adjustments, or patient follow-up initiatives.

#### Business Question 19
**Manager's Question**

Top 3 Doctors with Highest Cancellation Rate

```python
query = """
SELECT d.doctor_id , d.first_name, d.last_name, 
sum(CASE WHEN a.status='Cancelled' THEN 1 ELSE 0 END)*100.0 / count(a.appointment_id)  cancellation_rate
FROM doctors d 
LEFT JOIN appointments a 
ON d.doctor_id =a.doctor_id
GROUP BY  d.doctor_id , d.first_name, d.last_name
ORDER BY cancellation_rate DESC
LIMIT 3;
"""
pd.read_sql(query, conn)
```

**Observation** : 
The query identifies the three doctors with the highest percentage of cancelled appointments based on their total scheduled appointments.

**Business Insight** : 
A high cancellation rate may indicate scheduling conflicts, frequent rescheduling, or operational inefficiencies. Identifying these doctors allows hospital management to investigate underlying causes and implement strategies to improve appointment reliability and patient satisfaction.

### Section 5 — Billing and Revenue Analysis

Objective

Evaluate hospital revenue, payment methods, insurance contributions, and doctor-wise revenue generation using collected billing data.

#### Business Question 20
**Manager's Question**

What is the total revenue collected by the hospital?

```python
query = """
SELECT sum(amount)
FROM billing 
WHERE payment_status = 'Paid';
"""
pd.read_sql(query, conn)
```

**Observation** : 
The query calculates the total revenue from successfully paid bills.

**Business Insight** : 
Total collected revenue is one of the hospital's most important financial KPIs and reflects actual income generated.

#### Business Question 21
**Manager's Question**

Which payment methods are generating the most collected revenue?

```python
query = """
SELECT payment_method,SUM(amount) AS total_amount
FROM billing
WHERE payment_status='Paid'
GROUP BY payment_method
ORDER BY total_amount DESC;
"""
pd.read_sql(query, conn)
```

**Observation** : 
The query summarizes collected revenue for each payment method with credit card with highest.

**Business Insight** : 
Understanding payment method preferences helps optimize payment services and improve patient convenience.

#### Business Question 22
**Manager's Question**

Which insurance provider has generated the highest collected revenue?

```python
query = """
SELECT p.insurance_provider ,sum(b.amount) AS total_amount
FROM patients p
LEFT JOIN billing b
ON p.patient_id =b.pa
WHERE b.payment_status ='Paid'
GROUP BY p.insurance_provider 
ORDER BY total_amount DESC
LIMIT 1;
"""
pd.read_sql(query, conn)
```

**Observation** : 
The query identifies the revenue generated by insurance provider with max revenue.

**Business Insight** : 
This analysis highlights the hospital's most valuable insurance partnerships and supports strategic financial planning.

#### Business Question 23
**Manager's Question**

Which doctors generate the highest collected revenue?

```python
query = """
SELECT  d.doctor_id , d.first_name,d.last_name , SUM(b.amount) AS Total_amount
FROM doctors d
LEFT JOIN appointments a 
ON d.doctor_id = a.doctor_id
LEFT JOIN treatments t 
ON a.appointment_id = t.appointment_id
LEFT JOIN billing b 
ON t.treatment_id = b.treatment_id
WHERE b.payment_status='Paid'
GROUP BY d.doctor_id , d.first_name,d.last_name
ORDER BY Total_amount DESC;
"""
pd.read_sql(query, conn)
```

**Observation** : 
The query calculates the total revenue collected from successful (Paid) billing records for each doctor and ranks them in descending order.

**Business Insight** : 
Doctor-wise revenue analysis helps evaluate financial performance across the medical staff. This information supports strategic decision-making related to resource allocation, departmental planning, performance evaluation, and understanding which specialties or doctors contribute most to the hospital's revenue.

#### Business Question 24
**Manager's Question**

Which payment method is used the most by patients?


```python
query = """
SELECT payment_method, count(payment_method) AS Total_transaction
FROM billing
GROUP BY payment_method
ORDER BY Total_transaction DESC
LIMIT 1;
"""
pd.read_sql(query, conn)
```

**Observation** : 
The query identifies the payment method with the highest number of successful transactions.

**Business Insight** : 
Understanding the most preferred payment method helps the hospital prioritize secure and convenient payment options for patients.

#### Business Question 25
**Manager's Question**

What is the average revenue collected per successful payment?

```python
query = """
SELECT AVG(amount) AS average_revenue
FROM billing
WHERE payment_status='Paid';
"""
pd.read_sql(query, conn)
```

**Observation** : 
The query calculates the average amount collected from each successful payment transaction.

**Business Insight** : 
Monitoring the average payment value helps evaluate billing performance and provides a benchmark for future financial analysis.
